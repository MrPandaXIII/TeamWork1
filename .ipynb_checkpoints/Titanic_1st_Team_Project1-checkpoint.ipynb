### PipeLine

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 
train_model_names = []
train_accuracies = []
train_precisions = []
train_recalls = []
train_f1_scores = []
train_roc_aucs = []
train_results = []

test_model_names = []
test_accuracies = []
test_precisions = []
test_recalls = []
test_f1_scores = []
test_roc_aucs = []
test_results = []

best_models = {}
cv = KFold(n_splits=5, shuffle=True, random_state=0)

#  
for name, pipeline in models.items():
    print(f' {name}...')
    grid = GridSearchCV(pipeline, param_grids.get(name, {}), cv=cv, n_jobs=-1)
    grid.fit(X_train, y_train)
    
    best_models[name] = grid.best_estimator_

    y_train_pred = grid.predict(X_train)
    y_test_pred = grid.predict(X_test)

    if hasattr(grid, "predict_proba"):
        y_train_proba = grid.predict_proba(X_train)[:, 1]
        y_test_proba = grid.predict_proba(X_test)[:, 1]
    else:
        y_train_proba = np.zeros_like(y_train_pred, dtype=float)
        y_test_proba = np.zeros_like(y_test_pred, dtype=float)

    #   train
    acc_train = accuracy_score(y_train, y_train_pred)
    prec_train = precision_score(y_train, y_train_pred)
    rec_train = recall_score(y_train, y_train_pred)
    f1_train = f1_score(y_train, y_train_pred)
    roc_train = roc_auc_score(y_train, y_train_proba)

    train_model_names.append(name)
    train_accuracies.append(acc_train)
    train_precisions.append(prec_train)
    train_recalls.append(rec_train)
    train_f1_scores.append(f1_train)
    train_roc_aucs.append(roc_train)
    train_results.append((name, acc_train, prec_train, rec_train, f1_train, roc_train))

    #   test
    acc_test = accuracy_score(y_test, y_test_pred)
    prec_test = precision_score(y_test, y_test_pred)
    rec_test = recall_score(y_test, y_test_pred)
    f1_test = f1_score(y_test, y_test_pred)
    roc_test = roc_auc_score(y_test, y_test_proba)

    test_model_names.append(name)
    test_accuracies.append(acc_test)
    test_precisions.append(prec_test)
    test_recalls.append(rec_test)
    test_f1_scores.append(f1_test)
    test_roc_aucs.append(roc_test)
    test_results.append((name, acc_test, prec_test, rec_test, f1_test, roc_test, grid.best_params_))

    print(f"{name} | : acc={acc_train:.3f}, f1={f1_train:.3f} | : acc={acc_test:.3f}, f1={f1_test:.3f}")

# =========================================
# ========  ================
# =========================================

def plot_metric_comparison(metric_values, model_names, title, ax, color):
    sorted_idx = np.argsort(-np.array(metric_values))
    ax.barh(np.array(model_names)[sorted_idx], np.array(metric_values)[sorted_idx], color=color)
    ax.set_title(title)
    for i, v in enumerate(np.array(metric_values)[sorted_idx]):
        ax.text(v + 0.01, i, f"{v:.2f}", va='center')
    ax.set_xlim(0, 1)
    ax.invert_yaxis()
    ax.grid(True)

#   train
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

plot_metric_comparison(train_accuracies, train_model_names, "Train Accuracy", axes[0], 'green')
plot_metric_comparison(train_precisions, train_model_names, "Train Precision", axes[1], 'blue')
plot_metric_comparison(train_recalls, train_model_names, "Train Recall", axes[2], 'orange')
plot_metric_comparison(train_f1_scores, train_model_names, "Train F1 Score", axes[3], 'purple')
plot_metric_comparison(train_roc_aucs, train_model_names, "Train ROC AUC", axes[5], 'red')

fig.delaxes(axes[4])  #  
plt.suptitle("    ( )", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

#   test
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

plot_metric_comparison(test_accuracies, test_model_names, "Test Accuracy", axes[0], 'green')
plot_metric_comparison(test_precisions, test_model_names, "Test Precision", axes[1], 'blue')
plot_metric_comparison(test_recalls, test_model_names, "Test Recall", axes[2], 'orange')
plot_metric_comparison(test_f1_scores, test_model_names, "Test F1 Score", axes[3], 'purple')
plot_metric_comparison(test_roc_aucs, test_model_names, "Test ROC AUC", axes[5], 'red')

fig.delaxes(axes[4])
plt.suptitle("    ( )", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

#    Accuracy
best_idx = np.argmax(test_accuracies)
best_model_name, best_acc, best_precision, best_recall, best_f1, best_roc_auc, best_params = test_results[best_idx]

print(f"\n  ( Accuracy): {best_model_name}")
print(f"Accuracy: {best_acc:.3f}, Precision: {best_precision:.3f}, Recall: {best_recall:.3f}, "
      f"F1 Score: {best_f1:.3f}, ROC AUC: {best_roc_auc:.3f}")
print(f" : {best_params}")
```

```python
for name, model in best_models.items():
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    fig.suptitle(f'{name} - Confusion Matrices', fontsize=14)

    # Train
    y_train_pred = model.predict(X_train)
    ConfusionMatrixDisplay.from_predictions(
        y_train, y_train_pred, ax=axes[0], cmap='Greens'
    )
    axes[0].set_title("Train")

    # Test
    y_test_pred = model.predict(X_test)
    ConfusionMatrixDisplay.from_predictions(
        y_test, y_test_pred, ax=axes[1], cmap='Blues'
    )
    axes[1].set_title("Test")

    plt.tight_layout()
    plt.show()
```

```python
#     
report_rows = []

#  
for name, model in best_models.items():
    # 
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    #  classification_report   
    train_report = classification_report(y_train, y_train_pred, output_dict=True)
    test_report = classification_report(y_test, y_test_pred, output_dict=True)

    #      'weighted avg'
    for label in ['0', '1', 'weighted avg']:
        row = {
            'Model': name,
            'Class': label,
            'Train Precision': train_report[label]['precision'],
            'Train Recall': train_report[label]['recall'],
            'Train F1-Score': train_report[label]['f1-score'],
            'Test Precision': test_report[label]['precision'],
            'Test Recall': test_report[label]['recall'],
            'Test F1-Score': test_report[label]['f1-score'],
        }
        report_rows.append(row)

#  DataFrame   
report_df = pd.DataFrame(report_rows)
pd.set_option("display.max_rows", None)
display(report_df.round(3))
```
